```python
import pandas as pd

listings = pd.read_csv("listings.csv.gz")
neighbourhoods = pd.read_csv("neighbourhoods.csv")
```

```python
print(listings.columns)
print(neighbourhoods.columns)
```

```python
listings["neighbourhood_cleansed"] = listings["neighbourhood_cleansed"].str.strip().str.lower()
neighbourhoods["neighbourhood"] = neighbourhoods["neighbourhood"].str.strip().str.lower()
```

```python
barrios_listings = set(listings["neighbourhood_cleansed"].unique())
barrios_neigh = set(neighbourhoods["neighbourhood"].unique())

print("Barrios en listings:", len(barrios_listings))
print("Barrios en neighbourhoods:", len(barrios_neigh))
```
